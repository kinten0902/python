# 提出物評価と平均発現数（ラベル別）

```python
# 全体 a11:良い a12:普通 a13:悪い
# 具体 a21:良い a22:普通 a23:悪い
# 工夫 a31:良い a32:普通 a33:悪い
# 適切 a41:良い a42:普通 a43:悪い

all_g = all[::4,:]

a11 = all_g[(all_g[:, 0] == 3)]
a12 = all_g[(all_g[:, 0] == 2)]
a13 = all_g[(all_g[:, 0] == 1)]

a21 = all_g[(all_g[:, 1] == 3)]
a22 = all_g[(all_g[:, 1] == 2)]
a23 = all_g[(all_g[:, 1] == 1)]

a31 = all_g[(all_g[:, 2] == 3)]
a32 = all_g[(all_g[:, 2] == 2)]
a33 = all_g[(all_g[:, 2] == 1)]

a41 = all_g[(all_g[:, 3] == 3)]
a42 = all_g[(all_g[:, 3] == 2)]
a43 = all_g[(all_g[:, 3] == 1)]

index = [4,5,6,7,8,9,10,18]

print("全体：-------------")
print("良い：",[round(np.mean(a11[:, i]), 1) for i in index])
print("普通：",[round(np.mean(a12[:, i]), 1) for i in index])
print("悪い：",[round(np.mean(a13[:, i]), 1) for i in index])



print("具体：-------------")
print("良い：",[round(np.mean(a21[:, i]), 1) for i in index])
print("普通：",[round(np.mean(a22[:, i]), 1) for i in index])
print("悪い：",[round(np.mean(a23[:, i]), 1) for i in index])

print("工夫：-------------")
print("良い：",[round(np.mean(a31[:, i]), 1) for i in index])
print("普通：",[round(np.mean(a32[:, i]), 1) for i in index])
print("悪い：",[round(np.mean(a33[:, i]), 1) for i in index])

print("適切：-------------")
print("良い：",[round(np.mean(a41[:, i]), 1) for i in index])
print("普通：",[round(np.mean(a42[:, i]), 1) for i in index])
print("悪い：",[round(np.mean(a43[:, i]), 1) for i in index])

```

# 提出物評価と発言数との相関係数

```python
# 全体 y1
# 具体 y2
# 工夫 y3
# 適切 y4

all_g = all[::4,:]

index = [4,5,6,7,8,9,10,18]
y1 = all_g[:,0]
y2 = all_g[:,1]
y3 = all_g[:,2]
y4 = all_g[:,3]

print("全体：",[round(np.corrcoef(all_g[:, i], y1)[0,1],2) for i in index])
print("具体：",[round(np.corrcoef(all_g[:, i], y2)[0,1],2) for i in index])
print("工夫：",[round(np.corrcoef(all_g[:, i], y3)[0,1],2) for i in index])
print("適切：",[round(np.corrcoef(all_g[:, i], y4)[0,1],2) for i in index])


```

# 提出物評価と発言数の偏りとの相関係数

```python
test = all

index = [4,5,6,7,8,9,10,18]
for i in index:
    for g in range(46):
        aa = [test[g*4+1][i],test[g*4+2][i],test[g*4+3][i]]
        if(np.std(aa) == 0 or np.mean(aa) == 0):
            test[g*4][i] = 0.0
        else:
            test[g*4][i] = np.std(aa)/np.mean(aa)

all_g = all[::4,:]
# print(np.round(all_g,2))
y1 = all_g[:,0]
y2 = all_g[:,1]
y3 = all_g[:,2]
y4 = all_g[:,3]

print("全体：",[round(np.corrcoef(all_g[:, i], y1)[0,1],2) for i in index])
print("具体：",[round(np.corrcoef(all_g[:, i], y2)[0,1],2) for i in index])
print("工夫：",[round(np.corrcoef(all_g[:, i], y3)[0,1],2) for i in index])
print("適切：",[round(np.corrcoef(all_g[:, i], y4)[0,1],2) for i in index])
```
